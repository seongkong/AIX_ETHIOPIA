# Pitch Size

```python
import matplotlib.pyplot as plt

from mplsoccer import Pitch
```

## statsbomb

```python
pitch = Pitch(
    pitch_type="statsbomb",
    line_color="black",
    label=True, axis=True
)

fig, ax = pitch.draw()

plt.show()
```

```python
pitch.dim.length, pitch.dim.width
```
