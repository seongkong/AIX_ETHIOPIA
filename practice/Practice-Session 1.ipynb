# Pitch

### 1 - 피치를 그려라.
### 조건 corner arc 그리고, 채우기는 grass, line color는  white 스트라이프

```python
import matplotlib.pyplot as plt

from mplsoccer import Pitch
```

```python
pitch = Pitch(
    corner_arcs=True,
    pitch_type="statsbomb",
    pitch_color='grass',
    line_color='white',
    stripe=True
)

fig, ax = pitch.draw()

plt.show()
```

### 2 - 주어진 x,y 좌표를 활용해 arrow와 ball을 그려라
### 화살표는 투명하게
### 화살표 끝에 축구공을 그리기, 단 축구공이 위에 있어야함

```python
x1 = 103
x2 = 70
y1 = 14
y2 = 30
```

```python
fig, ax = pitch.draw()

pitch.arrows(x1, y1, x2, y2, ax=ax, color="gold", alpha=0.1)

pitch.scatter(x2, y2, ax=ax, marker="football", s=300, zorder=999)
pitch.lines(x1, y1, x2, y2, ax=ax, color="gold", comet=True)

plt.show()
```
